## <서답형2(단답형)>  추첨자 수와 학급당 학생 수를 입력받아 학생 번호 중 추첨자 수만큼을 무작위로 뽑고, 당첨되지 않은 번호 중에서 추가로 보너스 번호 1개를 뽑는 프로그램을 제작하고자 한다. 이때, (A)에 들어갈 내용을 쓰시오. (단, 리스트 함축(list comprehension)이나 filter() 함수를 사용하며, 매개변수 이름은 자유롭게 정해도 된다.)  [총 6.4점]

```python
import random

def lucky(n, nums) :
    winner = random.sample(range(1, nums+1), n)
    winner.sort()
    print(f'이번 주 행운 번호: {winner}')
    # non_winner = [i for i in range(1, nums+1) if i not in winner]
    non_winner = list(filter(lambda x:x not in winner, range(1, nums+1)))
    bonus = random.choice(non_winner)
    print(f'보너스 번호: {bonus}')

n, class_numbers = map(int, input('추첨자 수와 학급당 학생 수를 입력하세요(띄어쓰기로 구분): ').split())
lucky(n, class_numbers)
```

## <서답형4(단답형)>  <코드>를 실행했을 때, (1)~(5)의 출력 결과를 쓰시오.  [총 5.0점]

```python
import math
a = -1.5
b = 2.5

a, b = abs(a), abs(b)
print(a, b)              # (1) 1.5 2.5
a, b = round(a), round(b)
print(a, b)              # (2) 2 2
gm = math.sqrt(a * b)
print(gm)                # (3) 2.0
ft = math.factorial(a * b)
print(ft)                # (4) 24
lg = math.log(a * b, a)
print(lg)                # (5) 2.0
```

## <서답형5(단답형)>  특정 높이에서 물체를 가만히 놓아 자유 낙하시킬 때, 바닥에 도달하는 데 걸리는 시간과 도달 직전의 속도를 구하는 프로그램을 <조건>을 참고하여 작성하고자 한다. 이를 완성하기 위해 <코드>의 (1)~(3)에 들어갈 내용을 쓰시오.  [총 6.3점 / 각 2.1 점]

image.png

```python
import math

def free_fall( height, gravity=9.8 ) :
    time = math.sqrt(2 * height / gravity)
    velocity = gravity * time
    return  time, velocity, gravity

values = list(map(float, input('높이와 중력 가속도 입력: ').split()))
# print(*values)
t, v, g = free_fall( *values )
print(f'낙하 시간: {round(t, 2)}초')
print(f'도달 속도: {round(v, 2)}m/s')
print(f'중력 가속도: {g}')
```

## [서6~서7] 입력된 문자열이 회문인지 판별하는 프로그램을 제작하고자 한다. 다음 <코드>를 보고 물음에 답하시오.

image.png

```python
def Palindrome(text) :
    if len(text) <= 1 :
        return True
    if text[0] == text[-1] :
        return Palindrome(text[1:-1])
    else :
        return False

my_str = input('문자열을 입력하세요:')
if len(my_str) <= 1 :
    print('2글자 이상이어야 합니다.')
else :
    if Palindrome(my_str) :
        print('회문입니다!')
    else :
        print('회문이 아닙니다.')
```

## <서답형7(단답형)> <코드>에 있는 Palindrome 함수와 동일한 결과가 나오도록 <보기>의 Palindrome 함수를 완성하고자 한다. (A)에 들어갈 내용을 쓰시오.  [4.2점]

```python
def Palindrome(text) :
    my_list = list(text)
    my_rev = my_list[::-1]
    my_rev = list(reversed(my_list))
    return my_list == my_rev
```

```python
f = [0, 1]
n = int(input( ))
if n < 0 :
    print('0 이상의 정수여야 합니다.')
else :
    for i in range( 2, n+1 ):
        f.append(f[i-1] + f[i-2])
    print(f'{n}번째 피보나치 수: {f[n]}')
```
